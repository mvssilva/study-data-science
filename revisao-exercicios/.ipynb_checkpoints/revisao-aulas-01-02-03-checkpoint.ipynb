Tentando decifrar se uma senha de 4 digitos é boa ou não.

```python
def analisar_senhas(senha):
    senha = str(senha)

    if len(senha) != 4 or not senha.isdigit():
        return "ERRO - A senha deve conter exatamente 4 números!"

    if len(set(senha)) == 1:
        return "RUIM - Números repetidos são fáceis de prever"

    sequencia_crecente = "0123456789"
    if senha in sequencia_crecente:
        return "RUIM - Sequência numérica crescente detectada."

    sequencia_decrecente = "9876543210"
    if senha in sequencia_decrecente:
        return "RUIM - Sequência numérica decrescente detectada."

    padroes_teclado = ["0147", "7410", "0258", "8520", "1379", "2580", "1470", "3690", "1590", "7531"]
    if senha in padroes_teclado:
        return "RUIM - Padrão geométrico de teclado detectado."
    return "BOA - Não foram detectados padrões de repetição ou sequências."
```

```python
testes = ["mv12", "12345", "1111", "1234", "5432", "0147", "8139"]

for t in testes: 
    print(f"Senha {t}: {analisar_senhas(t)}")
```
